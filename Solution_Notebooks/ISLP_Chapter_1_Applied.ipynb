9 (c) What is the mean and standard deviation of each quantitative predictor?

```python
quantitative_predictors = ['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'year', 'origin']

# Calculate mean and standard deviation of each quantitative predictor in the "Auto_df" dataframe with rounded values
for predictor in quantitative_predictors:
    mean_value = round(Auto_df[predictor].mean(), 2)
    std_deviation = round(Auto_df[predictor].std(), 2)
    print(f"Predictor: {predictor}, Mean: {mean_value}, Standard Deviation: {std_deviation}")
```

9 (d) Now remove the 10th through 85th observations. What is the range, mean, and standard deviation of each predictor in the subset of the data that remains?

```python
# Remove the 10th through 85th observations
subset_df = Auto_df.drop(Auto_df.index[9:85])

# Calculate range, mean, and standard deviation of each predictor in the subset
quantitative_predictors = ['mpg', 'cylinders', 'displacement', 'weight', 'acceleration', 'year', 'origin']

for predictor in quantitative_predictors:
  predictor_range = subset_df[predictor].max() - subset_df[predictor].min()
  predictor_mean = round(subset_df[predictor].mean(),2)
  predictor_std = round(subset_df[predictor].std(),2)
  print(f"Predictor: {predictor}")
  print(f"Range: {predictor_range}")
  print(f"Mean: {predictor_mean}")
  print(f"Standard Deviation: {predictor_std}")
  print("\n")
```

```python
selected_features = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM']
sns.pairplot(boston_df[selected_features])
plt.show()
```

```python
selected_variables = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Calculate the correlation matrix
correlation_matrix = boston_df[selected_variables].corr()

# Print the correlation matrix
print("Correlation Matrix for Selected Variables:")
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", mask=mask)
plt.title("Lower Triangle Correlation Matrix Heatmap for Selected Variables")
plt.show()
```

As we see, the correlation between CRIM (Per Capita crime rate) and RAD (index of accessibility to radial highways) is quite high. So we can say that crimes are higher near the Highways.

10 (i) In this data set, how many of the suburbs average more than seven rooms per dwelling? More than eight rooms per dwelling? Comment on the suburbs that average more than eight rooms per dwelling.

```python
# Count of houses with average more than seven rooms per dwelling and more than eight rooms per dwelling
houses_above_7_rooms = boston_df[boston_df['RM'] > 7].shape[0]
houses_above_8_rooms = boston_df[boston_df['RM'] > 8].shape[0]

print("Houses with more than seven rooms per dwelling:", houses_above_7_rooms)
print("Houses with more than eight rooms per dwelling:", houses_above_8_rooms)
```

```python
# Filter the data where average rooms per dwelling is more than 8
filtered_data = boston_df[boston_df['RM'] > 8]

# Calculate the average per capita crime rate for the original data and the filtered data
avg_crim_original = boston_df['CRIM'].mean()
avg_crim_filtered = filtered_data['CRIM'].mean()

if avg_crim_original > avg_crim_filtered:
    print("Overall household has a higher per capita crime rate than households with more than 8 rooms")
else:
    print("Houses with more than eight rooms per dwelling has a higher per capita crime rate")
```
